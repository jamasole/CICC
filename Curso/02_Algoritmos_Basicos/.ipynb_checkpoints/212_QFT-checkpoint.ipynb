<b> LaTransformada de Fourier Cuántica </b>

La Transformada de Fourier Discreta (TFD), y puede considerarse una aplicación lineal sobre el espacio vectorial de dimension $N$ que transforma una vector de datos $x_i$ en otro $y_j$ dado por 

$$
y_k  = \sum_{j} \omega^{kj} x_j = \sum_{j}  \frac{1}{\sqrt{N}} e^{2\pi i  kj/N} x_j
$$ 

donde
$$
\omega  = \frac{1}{\sqrt{N}} e^{2\pi i/N}
$$

La *transformada de Fourier Cuántica* es la misma operación con una notación adaptada. Ahora $U^{TFC}_n$ será un  operador unitario.


Sean $\{\ket{x}\}$ con $x=0,...,N-1 = 2^n-1$ los elementos de la base computacional de $\Hil^{\otimes n}$ el espacio de Hilbert de n-qubits

La base   imagen del operador $U^{TFC}_n \equiv U_n$   define la *base de Fourier*  $\ket{\tilde x}$
donde se suele emplear una tilde (~) para referirse ella.

$$
U_n: \ket{x} \to {\ket{\tilde x}} =  \frac{1}{\sqrt{N}} e^{2\pi i  yx i/N} \ket{x} 
$$

donde $N=2^n$


Los elementos de matriz son evidentemte las fases

$$
\bra{x} U_n \ket{y} =  \frac{1}{\sqrt{N}} e^{2\pi i  x y/N} = \omega^{xy}
$$


 
$\bullet$ Para $n=1$,    la TFC no es otra que la puerta de Hadamard.
$$
U_1 = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}
$$

$\bullet$ Para $n=2 \to  e^{2\pi i/4} = i$ y entonces
$$
U_2 =\frac{1}{2}
\begin{bmatrix} 1 & 1 & 1 & 1 \\ 1 & -i & -1 & i \\ 1 & - 1 & 1 & - 1 \\ 1 & i & -1 & -i  \end{bmatrix}
$$

*Notación* : un número  $x\in {\mathbb Q} $ puede representarse en base 2

\begin{eqnarray}
x &=& x_{n-1} \cdots x_1 x_0\, .\, x_{-1}  \cdots x_{-m} \nonumber\\ \rule{0mm}{7mm}
&=&  x_{n-1}2^{n-1}   +   \cdots + x_1 2^1+ x_0 v    +  x_{-1}2^{-1} +  \cdots  +   x_{-m}2^{-m}\, .
\end{eqnarray}


en particular, if $x\in S_n =[0,...,2^n-1]$

\begin{eqnarray}
x/2&~=~& x_{n-1} 2^{n-2} + x_{n-1} 2^{n-3} + ... + x_1 + x_0/2  ~~=~~ x_{n-1}...x_1 + 0. x_0 \nonumber\\ \rule{0mm}{10mm}
x/2^2 &~=~& x_{n-1} 2^{n-3} + x_{n-1} 2^{n-4} + ... + x_1/2 + x_0/2^2  ~~=~~ x_{n-1}...x_2 + 0.x_1 x_0 \nonumber\\ \rule{0mm}{10mm}
&\vdots & \nonumber \\ \rule{0mm}{10mm}
x/2^n &~=~&  \frac{x_{n-1}}{2} + \frac{x_{n-2}}{2^2}  + ... +  \frac{x_{0}}{2^{n}} ~~=~~ 0.x_{n-1}x_{n-2}...x_0 \, .
\end{eqnarray}



Ahora vamos a estudiar la acción de $U^{\rm QFT}_n$ sobre un elemento $\ket{x} = \ket{x_{n-1}}\ket{x_{n-2}}... \ket{x_0}$
de la base computacional 

\begin{eqnarray}
 \ket{\tilde x}  ~\equiv ~  U_n^{\rm QFT}\ket{x}  &=&  \frac{1}{\sqrt{N}} \sum_{y=0}^{N-1} e^{2\pi i  xy/2^n}\ket{y} \label{TFC}
\nonumber\\ \rule{0mm}{9mm}
&=&  \frac{1}{\sqrt{N}} \sum_{y_1,...,y_n=\{0,1\}} e^{2\pi i x \left(\frac{y_{n-1}}{2} + \frac{y_{n-2}}{2^2}  + .. . + \frac{y_0}{2^n}\right) } \ket{y_{n-1}}\ket{y_{n-2}}....\ket{ y_0} \hspace{5cm}
\end{eqnarray}

\begin{eqnarray}
&=&  \frac{1}{\sqrt{N}} \left( \ket{0} + e^{2\pi i \frac{x}{2}}\ket{1}\right)\left( \ket{0} + e^{2\pi i \frac{x}{2^2}}\ket{1}\right)...\left( \ket{0} + e^{2\pi i \frac{x}{2^n}}\ket{1}\right)
\nonumber\\
%\pause
\rule{0mm}{9mm}
&=& \frac{1}{\sqrt{N}} \left( \ket{0} + e^{2\pi i \, 0.x_0}\ket{1}\right)\left( \ket{0} + e^{2\pi i \, 0.x_1x_0}\ket{1}\right)...\left( \ket{0} + e^{2\pi i \, 0.x_{n-1}...x_0}\ket{1}\right)   \nonumber \\
\rule{0mm}{9mm}
&\equiv & ~~ \ket{\tilde x_{n-1}}\ket{\tilde x_{n-2}}\cdots \ket{\tilde x_{0}}   \nonumber
\end{eqnarray}



 Hemos usado que, por ejemplo 
<br>
<br>
$$
e^{2\pi i x/2} = e^{2\pi i (x_{n-1}...x_1 + 0. x_0)}= e^{2\pi i (x_{n-1}...x_1)}e^{2\pi i \, 0. x_0} = e^{2\pi i\, 0.x_0}
$$
<br>
dado que $x_{n-1}...x_1$ es un número entero

## Circuito asociado a una TFC

El circuito asociado a una TCF sólo involucra la puerta de Hadamard $H$ y la puerta *puerta de fase discreta* $R_k \equiv P(\phi = \pi/2^k)$

$$
R_{k} = \begin{bmatrix} 1 & 0 \\ 0 & e^{ 2\pi i /2^{k}} \end{bmatrix}  
$$

que aparecerá de forma controlada $\cg{R_k}$.


Será util escribir la *acción* de $R_k$ en la forma siguiente
$$
(R_{k})^{x}\ket{y}  = e^{2 \pi i   \, x y/2^{k}}\ket{y}P_k =
$$

Veamos el circuito

Vamos a analizar la acción del primer bloque 

$$
(H\ket{x_{n-1}})\ket{x_{n-2}...x_0} =\left( \ket{0} + e^{2\pi i \left(\frac{x_{n-1}}{2}\right)} \ket{1}\right) \ket{x_{n-2}...x_0}
$$

$$
\left(\rule{0mm}{3mm} R_{2}^{x_{n-2}}H\ket{x_{n-1}}\right) \ket{x_{n-2}...x_0}  = 
 \left( \ket{0} + e^{2\pi i \left(\frac{x_{n-1}}{2} + \frac{x_{n-2}}{2^2}\right)} \ket{1}\right) \ket{x_{n-2}...x_0} 
$$

\begin{eqnarray}
\left(\rule{0mm}{5mm}R_{{(n-1)}}^{x_{0}} ... R_{3} ^{x_{n-3}}  R_{2}^{x_{n-2}}H\ket{x_{n-1}}\right) \ket{x_{n-2}...x_0}& =& 
\left( \ket{0} + e^{2\pi i \, 0.x_{n-1}\cdots x_0}  \ket{1} \right) \ket{x_{n-2}...x_0} \nonumber\\ \rule{0mm}{8mm}
& \equiv&  \ket{\tilde x_{0}} \ket{x_{n-2}... x_1x_0} \nonumber
\end{eqnarray}


El primer bloque ha generado el estado ecuatorial $\ket{\tilde x_0}$ pero ¡*en la posición equivocada*!

Si repetimos el mismo procedimiento con los siguientes cúbits $\ket{x_{n-2}}$ obtendremos finalmente

$$
 \ket{\tilde x_{0}}\ket{\tilde x_{1}}  ... \ket{\tilde{x}_{n-2}} \ket{\tilde x_{n-1}}
$$

que es *casi* el estado que estábamos buscando. 

La parte final del circuito introduce los operadores de SWAP que rectifican el orden de los cúbits 
<br>
<br>
$$
\hbox{SWAP}^{\otimes n} \,  (\ket{\tilde x_{0}}  ...  \ket{\tilde x_{n-1}} ) ~=~ 
\ket{\tilde x_{n-1}} ... \ket{\tilde{x}_{0}} ~~ \equiv ~~ \ket{\tilde x}
$$

```python

def qft(qc,n):
    """n-qubit QFT on q in circ."""
    for j in reversed(range(n)):
        qc.h(j)
        for k in range(j):
            qc.cp(np.pi/2**(j-k), k, j)
        qc.barrier()
    for j in range(n//2):
        qc.swap(j,n-j-1)
        
def qft_dagger(qc,n):
    """n-qubit QFT on q in circ."""
    for j in reversed(range(n//2)):
        qc.swap(j,n-j-1)            
    for j in range(n):
        for k in reversed(range(j)):
            qc.cp(-2*np.pi/2**(j-k+1), k, j)
        qc.h(j)
        qc.barrier()

def swap_registers(circuit, n):
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    return circuit
```

```python
# Create a circuit with quantum register and classical register
num_qubits=4

qr=QuantumRegister(num_qubits,name='qr')
cr=ClassicalRegister(num_qubits,name='cr')

# Create the circuit
qc=QuantumCircuit(qr,cr,name="QFT")


# Encode the input state
#qc.x(0)
#qc.x(2)
qc.h(range(num_qubits))

qc.barrier()
        
# qft
qft(qc,num_qubits)
qc.barrier()
#qft_dagger(qc,num_qubits)

#qc.measure(qr,cr)

qc.draw(output='mpl')
```

Measure the results. The QFT of the uniform state $\ket{+\ldots+}= \frac{1}{\sqrt{N}}\sum_{i} \ket{i}$ is the state $|0 \ldots 0>$

```python
from qiskit import Aer, execute
from qiskit.tools.visualization import plot_histogram, plot_bloch_multivector

S_simulator = Aer.get_backend("statevector_simulator")
statevector = execute(qc,backend=S_simulator).result().get_statevector()
plot_bloch_multivector(statevector)
```

```python
qc.measure(qr,cr)


M_simulator = Aer.get_backend('qasm_simulator')
counts = execute(qc,backend=M_simulator).result().get_counts()
plot_histogram(counts)
```

 Clásicamente, el algoritmo más eficiente para calcular
la TFD, (la FFT), precisa de ${\cal O}(n 2^n)$ por tanto hemos transformado un problema de tipo  $NP$ en uno de tipo $P$.

<b> Búsqueda de un Período </b>
</table>   

Una aplicación directa de la transformada de Fourier cuántica tiene que ver con funciones $f: S_n \to S_n$ periódicas. 


**Promesa:** $f: S_n \to S_n$ es suprayectiva con un sólo período $f(x+P) = f(x)$. 

**Problema**:  encontrar $P$ con un numero mínimo de consultas al oráculo.


Supondremos por simplificar, que $P$ divide  a $N=2^n$. Es decir, existe un entero $m$ tal que  $\Rightarrow m P=N$. 


Clásicamente necesitaríamos evaluar $f$ un número de veces del orden de $N/2$ para encontrar dicho período. En el paradigma cuántico, dicha evaluación se puede hacer de forma
simultánea en un solo paso



\begin{eqnarray}
\ket{\Psi} &=& U_f (W_n\otimes I_n) \ket{0}   \ket{0}  \nonumber\\
&=& \rule{0mm}{8mm}U_f   \frac{1}{\sqrt{N}}\sum_{x=0}^{N-1} \ket{x}\ket{0} \nonumber\\
&=&   \rule{0mm}{7mm} \frac{1}{\sqrt{N}}\sum_{x=0}^{N-1} \ket{x}\ket{f(x)}
\end{eqnarray}


Escribamos. $ x = kP +l$, with $l=0,...,P-1$ and $k=0,...,m-1$
 
\begin{eqnarray}
\ket{\Psi} &=& \rule{0mm}{7mm} \frac{1}{\sqrt{N}}  \sum_{l=0}^{P-1}\sum_{k=0}^{m-1}\ket{kP + l}\ket{f(kP+l)}\nonumber\\
&=& \rule{0mm}{7mm}  \frac{1}{\sqrt{N}} \sum_{l=0}^{P-1}\left(\sum_{k=0}^{m-1}\ket{kP + l}\right)\ket{f(l)}\nonumber\\
\end{eqnarray}


Donde la última igualdad se sigue aplicando la condición de periodicidad de $f$. 

Debido a esta propiedad, 
si una evaluación del segundo registro da como resultado $\ket{f(l_0)}$, el estado al que colapse el primer
registro es  una superposición normalizada 

$$
\ket{\Psi'} = \frac{1}{\sqrt{m}}  \sum_{k=0}^{m-1}\ket{l_0 + kP}\, .
$$

A continuación efectuamos sobre este estado una transformación de Fourier cuántica


\begin{eqnarray}
\ket{\Psi'} = U_{\rm TFC} \ket{\Psi'} &=& \frac{1}{\sqrt{N m}} \sum_{y=0}^{N-1}   \sum_{k=0}^{m-1}e^{+2\pi i (l_0 + kP)y/N}\ket{y} \nonumber\\
&=& \frac{1}{\sqrt{N m}} \sum_{y=0}^{N-1}   e^{+2\pi i l_0 y/N} \left( \sum_{k=0}^{m-1} e^{+2\pi i ky/m}\right) \ket{y}   \nonumber\\
&=& \frac{1}{\sqrt{N m}} \sum_{y=0}^{N-1}   e^{+2\pi i l_0 y/N}  \left( \rule{0mm}{3.5mm}m \delta_{y, qm}\right) \ket{y}    = \sqrt{ \frac{m}{N}}  \sum_{q=0}^{P-1} e^{+2\pi i l_0 qm/N}    \ket{qm}  \nonumber\\
&=&\sqrt{ \frac{1}{P} } \sum_{q=0}^{P-1} e^{+2\pi i l_0 q/P }    \ket{qN/P}  \nonumber
\end{eqnarray}




El resultado es una superposición homogénea de estados $\ket{y}$ con una probabilidad de ocurrencia 

$$
P(y) = \left\{ \begin{array}{cl} 1/P & \hbox{si} ~ y = q N/P \in \{ 0, N/P, 2N/P,..., (P-1)N/P \} \\ 0 & \rule{0mm}{6mm} \hbox{si} \end{array} \right.
$$



Si $c$ es el valor obtenido, $c= kN/P$ para algún $k$. En general será necesario repetir la prueba varias veces para averiguar $P$.



**Ejemplo**: Si $n=4\Rightarrow N=16$ sólo hay 3 posibilidades con $P=2,4$ y $8$. Para $P=2$ sólo pueden medirse valores $c\in\{0,8 ,16\}$. Con $P=4$
valores $c\in\{0,4,8,12,16\}$, y con $P=8$ valores $c\in\{ 0,2,4,6,8,10,12,14,16 \}$.
